```python
import pandas as pd

#check the dataset
train = pd.read_csv('/content/train_file.csv')
train.head()
```

```python
train_new = train.drop(['IDLink', 'Source','Topic','Facebook','GooglePlus','LinkedIn'], axis=1)
```

```python
#tokenization
import nltk
from nltk import word_tokenize
nltk.download('punkt')
nltk.download('wordnet')
```

```python
import re
def tokenize(txt):
    tokens= re.split('\W+', txt)
    return tokens
train_new['Headline_tokenized']= train_new['Headline'].apply(lambda x: tokenize(x.lower()))
```

```python
with open("/content/NRC-Emotion-Lexicon-Wordlevel-v0.92.txt","r") as f:
  lexicontext = f.readlines()
wordline = []
for line in lexicontext:
  wordline.append(line.split("\t"))
```

```python
train_new['Headline_tokenized'].head()
```

```python
newlex = {}
for entry in wordline:
  if entry[1] == "positive":
    if int(entry[2][0]) > 0:
      newlex[entry[0]] = 1
  elif entry[1] == "negative":
    if int(entry[2][0]) > 0:
      newlex[entry[0]] = -1
```

```python
for i in train_new.index:#これをもとにしてスコアを足せるようにする

  tokens = train_new.loc[i,'Headline_tokenized']
  wnl = nltk.WordNetLemmatizer()
  lemmalist = [wnl.lemmatize(t) for t in tokens]
  #print(lemmalist)

  score_lists = []
  score = 0.0
  for lemma in lemmalist:
    if lemma.lower() in newlex:
      #print(lemma)
       score += newlex[lemma.lower()]
       #print(lemma+str(score))
       score_lists.append([i, lemma, score])
  #print(score_lists)
```

```python
score_df = pd.DataFrame(score_lists, columns = ['number', 'word', 'score'])
score_df
```
